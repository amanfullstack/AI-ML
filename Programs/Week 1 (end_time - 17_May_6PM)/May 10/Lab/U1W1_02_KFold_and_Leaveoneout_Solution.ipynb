## K-Fold Cross Validation

KFold cross-validation is used to estimate model performance by splitting the training data into `K` parts. In each round, one part is used for validation and the remaining parts are used for training. The process is repeated `K` times and the scores are averaged.

This helps reduce the risk of relying on a single train/test split and gives a more stable performance estimate.

## Leave One Out

Leave One Out is a special form of cross-validation. In this method, each sample is used once as a test set while the remaining samples are used for training. The generalization error estimate is obtained by repeating this procedure for every sample and averaging the results.

## Importing Required Packages

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
```

### Prepare the model

```python
diabetes_data = pd.read_csv('https://cdn.iiith.talentsprint.com/aiml/Experiment_related_data/diabetes.csv')
diabetes_data.head()
```

Extract the features and labels from the diabetes data.

```python
x_data = diabetes_data.iloc[:, :-1].values
y_data = diabetes_data.iloc[:, -1].values
x_data.shape, y_data.shape
```

### Apply KFold

```python
def crossvalidation(data):
    scores_Test = []
    for train_index, test_index in data.split(x_data):
        # Split the data into train and test
        x_train, x_test = x_data[train_index], x_data[test_index]
        y_train, y_test  = y_data[train_index], y_data[test_index]

        # Create DecisionTree classifier object with hyper parameters
        decision_tree2 = DecisionTreeClassifier(max_depth=2)

        # Fit the data into the model
        decision_tree2.fit(x_train, y_train)
        scores_Test.append(decision_tree2.score(x_test, y_test))
    print("Average score of the Testing set %.2f"%np.mean(scores_Test))
```

```python
# Set the KFold module for 5 splits
kf = KFold(n_splits=5)

# crossvalidation returns the average score of the test data
crossvalidation(kf)
```

## Summary

KFold cross-validation helps estimate model performance more reliably than a single split. Leave One Out is another cross-validation strategy where one sample is held out at a time.
