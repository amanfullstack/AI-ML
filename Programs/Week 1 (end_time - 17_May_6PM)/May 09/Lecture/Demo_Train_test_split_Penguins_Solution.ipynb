## Dataset

The dataset used in this demo contains penguin measurements and categorical attributes. The target variable is `species`, and the input features are the remaining columns used for prediction.

### Description

The dataset consists of the below 7 columns:

- **species:** penguin species (Chinstrap, Adélie, or Gentoo)
- **culmen length & depth:** the culmen is the upper ridge of a bird's beak
- **flipper_length_mm:** flipper length
- **body_mass_g:** body mass
- **island:** island name (Dream, Torgersen, or Biscoe)
- **sex:** penguin sex

## Train-Test Split Evaluation

The train-test split is a technique for assessing a machine learning algorithm's performance.

The procedure involves dividing a dataset into two subsets. The first subset, known as the training dataset, is used to fit the model. The second subset is not used to train the model; instead, the model is fed the dataset's input element, and predictions are made and compared to expected values. The second dataset is known as the test dataset.

* Train Dataset: Used to fit the machine learning model.
* Test Dataset: Used to evaluate the fitted machine learning model.

The goal is to estimate the machine learning model's performance on new data that was not used to train the model.

## Import Required Packages

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Load the data from the original hosted URL
df = pd.read_csv('https://cdn.iiith.talentsprint.com/aiml/Experiment_related_data/penguins.csv')
df.head(5)
```

```python
# Get the shape of the dataset
df.shape
```

```python
df.columns
```

```python
# Store the data and target values in two separate variables x and y
x = df[['island', 'culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']]
y = df['species']
```

```python
x.shape, y.shape
```

```python
# Split the data into train and test sets in the ratio of 70:30
# i.e. 70% of the data is the train set and 30% of the data is the test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```
